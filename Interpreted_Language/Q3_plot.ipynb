```python
import numpy as np
import matplotlib.pyplot as plt

# Load data
data = np.loadtxt("scaling_results.txt")
m_values, n_values, times = data[:, 0], data[:, 1], data[:, 2]

# First plot: Execution time as a function of m (fixed n)
plt.figure()
for n in np.unique(n_values):
    mask = (n_values == n)
    plt.plot(m_values[mask], times[mask], marker='o', label=f"n={int(n)}")

plt.xlabel("Matrix Rows (m)")
plt.ylabel("Execution Time (s)")
plt.title("TSQR Scaling Performance (Varying m)")
plt.xscale("log")  # Keep log scale, suitable for large data range
plt.yscale("log")
plt.xticks([100, 500, 1000, 5000, 10000], labels=[100, 500, 1000, 5000, 10000])  # Ensure clear tick marks
plt.legend()
plt.grid()
plt.show()

# Second plot: Execution time as a function of n (fixed m)
plt.figure()
for m in np.unique(m_values):
    mask = (m_values == m)
    plt.plot(n_values[mask], times[mask], marker='o', label=f"m={int(m)}")

plt.xlabel("Matrix Columns (n)")
plt.ylabel("Execution Time (s)")
plt.title("TSQR Scaling Performance (Varying n)")
plt.xscale("linear")  # Use linear scale, as the range of n values is small
plt.yscale("log")  # Keep log scale for y-axis, suitable for time data
plt.xticks([5, 10, 20, 50], labels=[5, 10, 20, 50])  # Explicit n values
plt.legend()
plt.grid()
plt.show()
```
